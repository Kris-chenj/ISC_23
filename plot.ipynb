```python
import matplotlib.pyplot as plt

# CPU time in seconds
cpu_time = [2*60 + 38.40, 2*60 + 31.21, 2*60 + 13.77, 1*60 + 29.19,
            1*60 + 59.13, 1*60 + 34.72, 1*60 + 39.21, 1*60 + 46.74,
            2*60 + 10.93, 1*60 + 58.66]

# WALL time in seconds
wall_time = [2*60 + 43.32, 2*60 + 34.08, 2*60 + 27.55, 1*60 + 42.93,
             2*60 + 8.06, 1*60 + 43.47, 1*60 + 45.77, 1*60 + 53.84,
             2*60 + 18.26, 2*60 + 2.01]

# number of processors
p = [128, 129, 192, 256, 384, 385, 416, 448, 480, 512]


# plot CPU and WALL time as a function of the number of processors
fig, ax = plt.subplots(figsize=(25, 5))
ax.plot(p, cpu_time, '-o', label='CPU time')
ax.plot(p, wall_time, '-o', label='WALL time')
ax.set_xlabel('Number of Processors')
ax.set_ylabel('Time (s)')
ax.set_title('Bridge CP input')
ax.legend()

# set the x-axis tick marks every 1
ax.set_xticks(range(min(p), max(p)+1, 20))

plt.show()
```
